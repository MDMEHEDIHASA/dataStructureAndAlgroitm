```python
import math
def findValue(matrix,target):
    visited = []
    for i in range(0,len(matrix)):
        for j in range(0,len(matrix[i])):
            visited.append(matrix[i][j])
    print(visited)
    def binarySearch(arr=visited,value=target):
        
        left = 0
        right = len(arr)-1
        middle = math.floor((left+right)/2)
        while(left<=right):
            if(arr[middle] ==value):
                return True
            if(arr[middle] >value):
                right = middle -1
            else:
                left = middle+1
            middle = math.floor((left+right)/2)
        return False
    return binarySearch()
```

```python
findValue([[1,3,5,7],[10,11,16,20],[23,30,34,60]],3)
```

```python
def firstUniqChar(s):
    visited = {}
    for i in range(0,len(s)):
        if s[i] in visited:
            visited[s[i]]+=1
        else:
            visited[s[i]] = 1
    print(visited)
    for item in visited:
        if(visited[item] == 1):
            return s.find(item)
    return -1
```

```python
def stairs(n):
    visited = [0 for i in range(0,n+1)]
    for i in reversed(range(0,n+1)):
        if i == n:
            visited[i] = 1
            continue;
        if i == n-1:
            visited[i] = 1
            continue
        visited[i] = visited[i+1] + visited[i+2]
    print(visited)
```

```python
stairs(4)
```
